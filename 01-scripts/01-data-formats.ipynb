```python
# Import Packages
import pandas as pd
import numpy as np
import os
import matplotlib.pyplot as plt
```

```python
# Download and open .csv file using Pandas
avg_monthly_precip = pd.read_csv("https://ndownloader.figshare.com/files/12710618")

# View the data downloaded and opened
avg_monthly_precip
```

```python
avg_monthly_precip.info()
```

```python
# Plot the data
fig, ax = plt.subplots()
avg_monthly_precip.plot(
    x="months", y="precip", title="Precipitation (mm) for One Year", ax=ax
)
ax.set(xlabel="Month", ylabel="Precipitation (mm)")
plt.show()
```

```python
## Open Miami Temperature data from the url
miami_temp_url = "https://www.ncdc.noaa.gov/cag/city/time-series/USW00012839-tmax-12-12-1895-2020.csv"

miami_temp = pd.read_csv(miami_temp_url)
miami_temp
```

```python
# Open Temperature skippinkg 3 rows and setting no data values
miami_temp = pd.read_csv(miami_temp_url, skiprows=3, na_values=-99)

# View the first 5 rows of the data
miami_temp.head()
```

# Challenge 5
Copy the code below into your code. Run the code. It should download and open a new dataset that has CO2 emissions for a field site in Barrow, Alaska (brw).

The code below will download your data into your working directory. You should be able to open it using the filename co2-emissions-barrow.csv.

```python
# Download greenhouse gas CO2 data
import urllib.request

greenhouse_gas_url = "https://gml.noaa.gov/aftp/data/trace_gases/co2/in-situ/surface/txt/co2_brw_surface-insitu_1_ccgg_MonthlyData.txt"

urllib.request.urlretrieve(url=greenhouse_gas_url, filename="co2-emissions-barrow.csv")
```

```python
df = pd.read_csv(
    "co2-emissions-barrow.csv", skiprows=157, sep="\s+", na_values=[-999.99, -9, -99.99]
)
df.head()
```

```python
df.info()
```

```python
fig, ax = plt.subplots()
df.plot(x="time_decimal", y="value", ax=ax, title="CO2 - Barrow, AK")
ax.set(xlabel="Date", ylabel="CO2")
```
